```python
def count_licks_in_trial(row, data_licks, t = 'laser_ts', t_window = 600):
    # Get licks within the start and end time of the trial
    licks_in_trial = len(data_licks[(data_licks[0] >= row[t]) & 
                                 (data_licks[0] <= row[t] + t_window)]) 
    return licks_in_trial
```

```python

def calc_d_prime(hit_rate, false_alarm_rate):
    # Ensure hit_rate and false_alarm_rate are within range to avoid infinity
    hit_rate = np.clip(hit_rate, 0.0001, 0.9999)
    false_alarm_rate = np.clip(false_alarm_rate, 0.0001, 0.9999)
    
    # Calculate Z scores
    z_hit_rate = norm.ppf(hit_rate)
    z_false_alarm_rate = norm.ppf(false_alarm_rate)
    
    # Calculate d-prime
    d_pr = z_hit_rate - z_false_alarm_rate
    
    return d_pr
```

```python
prev_same = []
prev_diff = []
prev_same_L = []
prev_diff_L = []
prev_same_R = []
prev_diff_R = []
corrects_L = []
corrects_R = []
mouse_ids = []
d_primes = []
for exp in data_all.keys():
    mouse_id = exp[:4]
    data = data_all[exp]['behav']
 #   print(exp)
    prev_same_correct_L = data[(data['choice']==1) & (data['side']==1) & (data['prev_reward']==1) ].shape[0] / (
                        data[(data['side']==1) & (data['prev_reward']==1) & (data['missed']==0) ].shape[0])
    prev_diff_correct_L = data[(data['choice']==1) & (data['side']==1) & (data['prev_reward']==-1) ].shape[0] / (
                        data[(data['side']==1) & (data['prev_reward']==-1) & (data['missed']==0)].shape[0])
    

  #  print('left', round(prev_same_correct,2),round(prev_diff_correct,2))
    prev_same_correct_R = (data[(data['choice']==-1) & (data['side']==-1) & (data['prev_reward']==-1) ].shape[0]) / (
                        data[(data['side']==-1) & (data['prev_reward']==-1) & (data['missed']==0)].shape[0])
    prev_diff_correct_R = data[(data['choice']==-1) & (data['side']==-1) & (data['prev_reward']==1) ].shape[0] / (
                       data[(data['side']==-1) & (data['prev_reward']==1) & (data['missed']==0)].shape[0])
    # print(round(prev_same_correct_R,2),round(prev_diff_correct_R,2))
    
    

    # overall, for both same and diff
    correct_L = (data[(data['choice']==1) & (data['side']==1)].shape[0]) / (
                        data[(data['side']==1) & (data['missed']==0)].shape[0])
    correct_R = data[(data['choice']==-1) & (data['side']==-1)].shape[0] / (
                       data[(data['side']==-1) & (data['missed']==0)].shape[0])

    d_pr = calc_d_prime(correct_L, 1 - correct_R)

    # append everything only if things look ok
    cutoff = 0
    if (prev_same_correct_L > cutoff) & (prev_same_correct_R > cutoff) & \
       (prev_diff_correct_L > cutoff) & (prev_diff_correct_R > cutoff):
        prev_same.append(prev_same_correct_L)
        prev_diff.append(prev_diff_correct_L)
        prev_same_L.append(prev_same_correct_L)
        prev_diff_L.append(prev_diff_correct_L)
        prev_same.append(prev_same_correct_R)
        prev_diff.append(prev_diff_correct_R)
        prev_same_R.append(prev_same_correct_R)
        prev_diff_R.append(prev_diff_correct_R)
        corrects_L.append(correct_L)
        corrects_R.append(correct_R)
        d_primes.append(d_pr)
        d_primes.append(d_pr)
        mouse_ids.append(mouse_id)
        mouse_ids.append(mouse_id)
        mouse_ids.append(mouse_id)
        mouse_ids.append(mouse_id)
```

```python
def plot_lick_row(data, t_pre, t_post, y, prev='all', choice='all', subplot=False,ax_row=0,ax_col=0):
    filtered_data = data['behav']
    data_licks_L = data['licks_L']
    data_licks_R = data['licks_R']

    if prev != 'all':
        if prev == 'left':
            prev_type = 1  
        elif prev == 'right':
            prev_type = -1
        elif prev == 'no':
            prev_type = 0
        filtered_data = filtered_data[filtered_data['prev_reward'] == prev_type]

    if choice != 'all':
        if choice == 'left':
            choice_type = 1  
        elif choice == 'right':
            choice_type = -1
        elif choice == 'no':
            choice_type = 0
        filtered_data = filtered_data[filtered_data['choice'] == choice_type]
    
    for index, row in filtered_data.iterrows():
        # Find licks within 4000 ms after tone 
        tone_time = row['tone_ts']
        licks_L = data_licks_L[(data_licks_L[0] >= tone_time-t_pre) & (data_licks_L[0] <= tone_time + t_post)]
        licks_R = data_licks_R[(data_licks_R[0] >= tone_time-t_pre) & (data_licks_R[0] <= tone_time + t_post)]
        licks_L = licks_L[0].tolist()
        licks_R = licks_R[0].tolist()
        
        #if len(licks_L) > 0:
        ys_L = [y]*len(licks_L)
        ys_R = [y]*len(licks_R)
        licks_L = [x - tone_time for x in licks_L]
        licks_R = [x - tone_time for x in licks_R]
        if subplot:
            axs[ax_row].plot(licks_L, ys_L,'|',color = 'b')
            axs[ax_row].plot(licks_R, ys_R,'|',color = 'r')
        else:
            plt.plot(licks_L, ys_L,'|',color = 'b')
            plt.plot(licks_R, ys_R,'|',color = 'r')
        y -= 1

    return y, licks_L, licks_R
```
